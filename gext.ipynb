```python
import ast

df["genre_names"] = df["genres"].apply(
    lambda x: [d["name"] for d in ast.literal_eval(x)] if x != "[]" else []
)

df["genre_names"]
```

```python
def extract_genre_names(x):
    try:
        genres = ast.literal_eval(x)  # convert string to list of dicts
        if isinstance(genres, list):
            return ", ".join(sorted(set(d["name"].strip().lower() for d in genres if "name" in d)))
        else:
            return ""
    except Exception:
        return ""

df["genre_names"] = df["genres"].apply(extract_genre_names)
```
